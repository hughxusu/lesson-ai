```python
import pandas as pd

df = pd.read_csv('area_price_data.csv', index_col='Index')
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['Gr Liv Area'], df['SalePrice'], alpha=0.7, color='blue')
plt.xlabel('Gr Liv Area (sq ft)')
plt.ylabel('Sale Price ($)')
plt.title('Scatter Plot: Gr Liv Area vs Sale Price') 
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
